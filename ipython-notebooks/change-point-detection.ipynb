```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
# first make some sample data
# we use sin function because we know it will cross zero which is
# common change point
x = np.linspace(0, 100, 10000)
y = np.sin(x)

fig, ax = plt.subplots(figsize=(15, 5))
ax.plot(x, y)
```

```python
# we can use numpy sign function which tells you what the sign of each element
# of an array is
# https://numpy.org/doc/stable/reference/generated/numpy.sign.html
sign_change = np.sign(y)

fig, ax = plt.subplots(figsize=(15, 5))
ax.plot(x, y)
ax.plot(x, sign_change)
```

```python
# taking the difference between each element n and element n-1 tells when
# the sign changes
diff_sc = np.diff(sign_change)

fig, ax = plt.subplots(figsize=(15, 5))
ax.plot(x, y)
ax.plot(x[:-1], diff_sc)
```

```python
# detecting only the positive change points we use the numpy where function
# https://numpy.org/doc/stable/reference/generated/numpy.where.html
positive_change_point = np.where(diff_sc > 1)[0]

fig, ax = plt.subplots(figsize=(15, 5))
ax.plot(x, y)
ax.plot(x[positive_change_point], y[positive_change_point], marker='^', color='red', linewidth=0, markersize=10)
```

```python
# the entire solution can be found here.

x = np.linspace(0, 100, 10000)
y = np.sin(x)

fig, ax = plt.subplots(figsize=(15, 5))
ax.plot(x, y)

sign_change = np.sign(y)

ax.plot(x, sign_change)

diff_sc = np.diff(sign_change)

ax.plot(x[:-1], diff_sc)

positive_change_point = np.where(diff_sc > 1)[0]

ax.plot(x[positive_change_point], y[positive_change_point], marker='^', color='red', linewidth=0, markersize=10)

ax.plot([-100, 200], [0, 0], color='black', alpha=0.5)
ax.set_xlim(-5, 105)

print('number of positive change points:', str(len(positive_change_point)))
```
