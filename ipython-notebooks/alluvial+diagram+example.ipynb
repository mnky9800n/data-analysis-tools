# Making alluvial diagrams in R

This notebook shows how to make an alluvial diagram in R. You can find the full documentation here:

https://cran.rstudio.com/web/packages/alluvial/alluvial.pdf

But I have included some description so that this notebook is more useful.

```r
install.packages("alluvial") # install the library if you haven't already done so
```

```r
library(alluvial) # import the library
```

```r
tit <- as.data.frame(Titanic, stringAsFactors=False) # import some data
                                                     # in this case we use a sample dataset from the 
                                                     # Titanic passenger manifest.
```

```r
head(tit)
```

```r
alluvial(tit[,1:4]
         , freq=tit$Freq # numeric, vector of frequencies of the same length as the number of observations
         ,col = ifelse(tit$Survived == "Yes", "orange", "grey") # vector of colors of the stripes
         ,border = ifelse(tit$Survived == "Yes", "orange", "grey") # vector of border colors for the stripes
         ,hide = tit$Freq == 0 # logical, should particular stripe be plotted
         ,cex = 0.7 #  numeric, scaling of fonts of category labels and axis labels respectively.
)
```
