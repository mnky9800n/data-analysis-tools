```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
# create some data to plot
x = np.random.normal(0, 1, (10000, 2))

# create a 2x2 matrix of axes that use the same x and y limits
fig, ax = plt.subplots(2, 2, sharex=True, sharey=True, figsize=(10, 10))

def plot_hist(ax):
    '''
    A function to plot a 2d histogram
    
    This mainly exists to make it easier to edit all 4 plots at once
    
    hist2d returns all of the histogram information and the updated 
    axes object. this function only returns the updated axes object.
    '''
    x1, x2, x3, cbar = ax.hist2d(x[:,0], x[:,1], bins=50, cmap='plasma')
    return cbar

plot_hist(ax[0][0])
plot_hist(ax[0][1])
plot_hist(ax[1][0])

# returns the quadmesh axis that is necessary to know informations about
# the colormaps. see above function
cbar = plot_hist(ax[1][1])


# This is where you add the colorbar
# This creates a new axes that floats wherevere you put it
# It will not be dynamically resized. If you use tight_layout()
# then you will need to change where the float goes because it 
# will not dynamically move itself based on the tight_layout()
# specifications. This is why a UserWarning is thrown.
# The add_axes function takes a list of coordinates for where the 
# new axes is supposed to go with the dimensions [left, bottom
# , width, height] of the new Axes. All quantities are in 
# fractions of figure width and height.
cbar_ax = fig.add_axes([1.0085, 0.15, 0.05, 0.7])
fig.colorbar(cbar, cax=cbar_ax, label='value count')

fig.tight_layout()
```
