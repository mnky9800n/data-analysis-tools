```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.random.normal(0, 1, (10000, 2))

fig, ax = plt.subplots(2, 2, sharex=True, sharey=True, figsize=(10, 10))

def plot_hist(ax):
    x1, x2, x3, cbar = ax.hist2d(x[:,0], x[:,1], bins=50, cmap='plasma')
    return cbar

plot_hist(ax[0][0])
plot_hist(ax[0][1])
plot_hist(ax[1][0])
cbar = plot_hist(ax[1][1])

fig.subplots_adjust(right=10.8)
cbar_ax = fig.add_axes([1.0085, 0.15, 0.05, 0.7])
fig.colorbar(cbar, cax=cbar_ax, label='value count')

fig.tight_layout()
```
