# Create Random Data sets for Two Populations

1. Start with two data sets.
2. Create a new column called “target” in each data sets, and assign an integer value.
3. Concat the two data sets.
4. Provide a color to the color wheel look up dictionary.
5. Plot.

```python
df1 = pd.DataFrame(np.random.randint(150, size=(100,4))
                   , columns=['A', 'B', 'C', 'D'])
df1['target'] = 1

df2 = df1.copy()
df2.A = df2.A.apply(lambda x: 2.0*x)
df2.B = df2.B.apply(lambda x: np.exp(x))
df2.C = df2.C.apply(lambda x: x*x)
df2.D = df2.D.apply(lambda x: np.power(x, 0.5))

df2['target'] = 2

data = pd.concat([df1, df2]).reset_index()

del data['index']
```

### Example from Stack Overflow

http://stackoverflow.com/questions/21131707/multiple-data-in-scatter-matrix

```python
from pandas.tools.plotting import scatter_matrix
from sklearn import datasets

iris = datasets.load_iris()
iris_data = pd.DataFrame(data=iris['data'], columns=iris['feature_names'])
iris_data['target'] = iris['target']

color_wheel = {1: 'red'
               ,2:'green'
               ,3:'blue'}
colors = iris_data['target'].map(lambda x: color_wheel.get(x+1))
ax = scatter_matrix(iris_data, color=colors, alpha=0.6
                    ,figsize=(15,15), diagonal='hist')
```

```python
from pandas.tools.plotting import scatter_matrix

color_wheel = {1: 'navy'
               , 2: 'gold'
               , 3: 'crimson'}
colors = data['target'].map(lambda x: color_wheel.get(x))

ax = scatter_matrix(data, color=colors, alpha=0.6, figsize=(15,15)
                    ,diagonal='kde')
```
